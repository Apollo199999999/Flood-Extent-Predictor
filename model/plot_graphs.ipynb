# Plotting training and validation loss curves

## 0. Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
# Make sure scienceplots is installed
import scienceplots

plt.style.use(['science','ieee'])
```

## 1. Read data

```python
base_unet_losses = pd.read_csv("./output/base_unet_losses.csv")
base_vit_losses = pd.read_csv("./output/base_vit_losses.csv")
flood_vit_losses = pd.read_csv("./output/flood_vit_losses.csv")
```

```python
base_unet_losses.head()
```

## 2. Plot train losses

```python
plt.plot(flood_vit_losses["epoch"], flood_vit_losses["train_loss"], label="FloodViT")
plt.plot(base_unet_losses["epoch"], base_unet_losses["train_loss"], label="U-Net")
plt.plot(base_vit_losses["epoch"], base_vit_losses["train_loss"], label="BaseViT")
plt.legend()
plt.title("Training losses")
plt.xlabel("Epoch")
plt.ylabel("Loss")
```

## 2. Plot validation losses

```python
plt.plot(flood_vit_losses["epoch"], flood_vit_losses["val_loss"], label="FloodViT")
plt.plot(base_unet_losses["epoch"], base_unet_losses["val_loss"], label="U-Net")
plt.plot(base_vit_losses["epoch"], base_vit_losses["val_loss"], label="BaseViT")
plt.legend()
plt.title("Validation losses")
plt.xlabel("Epoch")
plt.ylabel("Loss")
```
