This notebook pushes the raw TIFF files downloaded and the generated tiles to HuggingFace

```python
from huggingface_hub import HfApi

api = HfApi()
```

## 1. Upload TIFFs to HuggingFace

```python
# Change folder_path as necessary
api.upload_large_folder(
    repo_id="MatthiasWang/geotiffs",
    repo_type="dataset",
    folder_path="E:\\RSI09\\v4\\tiff",
)
```

## 2. Upload processed tiles and metadata

Make sure you have the latest version of `huggingface_hub` with `hf-xet` installed.

```python
import os

# Change directory as needed
directory = "E:\\RSI09\\v4"
extensions = [".jsonl", ".json", ".npy"]

# Get all jsonl, json, npy files in directory
for path in os.listdir(directory):
        full_path = os.path.join(directory, path)

        if os.path.isfile(full_path) == False:
            continue
        
        # Get extension
        ext = os.path.splitext(full_path)[1]

        if ext not in extensions:
              continue
        
        # Upload file to HF
        api.upload_file(
            path_or_fileobj=full_path,
            path_in_repo=os.path.basename(full_path),
            repo_id="MatthiasWang/processed-tiles",
            repo_type="dataset",
        )
        
```
